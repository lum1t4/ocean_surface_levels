```python
import xarray as xr


"""
EUR : European Area (30.0625°W-42.0625°E ; 19.9375°N-66.0625°N)
Each pixel corresponds to 0.0625° x 0.0625° grid cell

Canary Islands: 
`--min-lon -40 --max-lon -10 --min-lat 22 --max-lat 29.5` equals to 120 x 321 since
witdh/longitude: (-10 - (-30.0625)) / 0.0625 = 20.0625 / 0.625 = 321 # minimum longitude in the dataset (EUR) is -30.0625
height/latitude: (29.5 - 22) / 0.0625 = 7.5 / 0.625 = 120

The full resolution dataset is 738 x 1154 pixels, which corresponds to the full EUR area.

To get a 224x224 for ViViTModel
224 * 0.0625 = 14 degrees
--min-lon -30.0625 --max-lon -16.0625
--min-lat 19.9375 --max-lat 33.9375
"""

# Calculate total size of the dataset in GB
def dataset_size(h: int = 738, w: int = 1154, c: int = 3, d: int = 11487):
    """
    Calculate the size of the dataset in GB.
    param h: height of the dataset in pixels
    param w: width of the dataset in pixels
    param c: number of channels (3 for RGB)
    param d: number of days in the dataset
    return: size of the dataset in GB
    """
    return h * w * c * d * 4 / (1024 ** 3)  # GB

print(f"Dataset size: {dataset_size():.2f} GB")  # Dataset size in GB
print(f"Dataset size: {dataset_size(120, 321, 3, 11487):.2f} GB")  # Dataset size in GB
print(f"Dataset size: {dataset_size(320, 320, 3, 11487 * 2):.2f} GB")  # Dataset size in GB
print(f"Dataset size: {dataset_size(224, 224, 3, 11487 * 2):.2f} GB")  # Dataset size in GB

OLD = "../data/raw/cmems_obs-sl_eur_phy-ssh_nrt_allsat-l4-duacs-0.125deg_P1D.nc"
NEW = "../data/raw/cmems_obs-sl_eur_phy-ssh_my_allsat-l4-duacs-0.0625deg_P1D.nc"
data = xr.open_dataset(NEW)

```
